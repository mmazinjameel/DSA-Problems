```python
from collections import defaultdict

class Solution:
    def countBalanced(self, arr):
        def get_balance(s):
            vowels = set('aeiou')
            v = sum(1 for ch in s if ch in vowels)
            c = len(s) - v
            return v - c  # delta: +1 per vowel, -1 per consonant

        balance = 0
        freq = defaultdict(int)
        freq[0] = 1  # To count balance-zero prefix from start
        result = 0

        for s in arr:
            balance += get_balance(s)
            result += freq[balance]
            freq[balance] += 1

        return result
```
