```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        res = ("", 0)  # (longest_palindrome, length)
        
        for i in range(len(s)):
            # Odd-length palindrome
            l, r = i, i  
            while l >= 0 and r < len(s) and s[l] == s[r]:
                if r - l + 1 > res[1]:
                    res = (s[l:r+1], r - l + 1)
                l -= 1
                r += 1

            # Even-length palindrome
            l, r = i, i + 1  
            while l >= 0 and r < len(s) and s[l] == s[r]:
                if r - l + 1 > res[1]:
                    res = (s[l:r+1], r - l + 1)
                l -= 1
                r += 1

        return res[0]  # Return the longest palindromic substring
```
