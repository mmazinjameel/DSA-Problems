```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        def helper(root):
            if not root:
                return

            if root == q or root == p:
                return root
            
            left = helper(root.left)
            right = helper(root.right)

            if left and right:
                return root

            return left if left else right


        return helper(root)  
```
