```python
from typing import List

class Solution:
    def findChampion(self, n: int, edges: List[List[int]]) -> int:
        in_degree = [0] * n

        for edge in edges:
            in_degree[edge[1]] += 1

        champion = -1

        for i in range(n):
            if in_degree[i] == 0:
                if champion == -1:
                    champion = i
                else:
                    return -1  # More than one node with in-degree 0 → no single champion

        return champion  # Return the champion if exactly one exists
```
