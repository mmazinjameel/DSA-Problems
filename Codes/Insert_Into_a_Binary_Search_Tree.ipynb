```python
class Solution:
    def insertIntoBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:
        
        def helper(node):
            if node is None:
                return TreeNode(val)  # Create new node when reaching a leaf
            
            if node.val > val:
                node.left = helper(node.left)  # Recur to the left subtree
            else:
                node.right = helper(node.right)  # Recur to the right subtree

            return node  # Return the modified tree
        
        return helper(root)
```
