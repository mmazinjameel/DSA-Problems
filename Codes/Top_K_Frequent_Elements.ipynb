```python
class Solution {
public:
    vector<int> topKFrequent(vector<int>& nums, int k) {
            // Create that freq hashmap
    unordered_map <int,int> freq_map;

    for (int n: nums){
        freq_map[n]++;
    }

    

    int max_freq = 0;
    for (const auto& pair : freq_map) {
        max_freq = max(max_freq, pair.second);
    }
    vector<vector<int>> buckets(max_freq + 1);

    // Populate the bucket
    for (const auto& pair: freq_map){
        buckets[pair.second].push_back(pair.first); 
    }


    // Extract out the elements


        vector<int> res;
        int index = 0;

        for (int i = max_freq; i >= 0; i--) {
            if (!buckets[i].empty()) {
                for (const int& n : buckets[i]) {
                    res.push_back(n);
                    index++;
                    if (index == k) {
                        return res; // Return when we have found the k most frequent elements
                    }
                }
            }
        }

        return res;
    }
};
```
