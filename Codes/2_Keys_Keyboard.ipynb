# $O(n^2)$


```python
class Solution:
    def minSteps(self, n: int) -> int:
        res = 0
        d = 2
        while n > 1:
            while n % d == 0:
                res += d
                n //= d
            d += 1
        return res
```

# Dynamic Programming

```python
class Solution:
    def minSteps(self, n: int) -> int:
        dp = [0] * (n + 1)  # dp[i] will hold min steps to get i 'A's

        for i in range(2, n + 1):
            dp[i] = i  # Worst case: Paste 1 at a time (1 + 1 + 1...)
            for j in range(1, i // 2 + 1):
                if i % j == 0:
                    # If j divides i, we can build j 'A's, then copy and paste
                    steps = dp[j] + (i // j)
                    dp[i] = min(dp[i], steps)
        
        return dp[n]
```
