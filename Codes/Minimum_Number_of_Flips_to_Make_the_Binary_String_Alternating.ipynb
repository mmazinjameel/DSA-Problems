```python
class Solution:
    def minFlips(self, s: str) -> int:
        n = len(s)
        s = s + s  # Extend the string to handle rotations
        alt1, alt2 = "", ""  

        # Generate two alternating pattern strings
        for i in range(2 * n):
            if i % 2 == 0:
                alt1 += "1"
                alt2 += "0"
            else:
                alt1 += "0"
                alt2 += "1"

        mismatch1, mismatch2 = 0, 0  # Count mismatches for alt1 and alt2
        min_flips = float("inf")

        # Initial mismatch count for the first window
        for i in range(n):
            if s[i] != alt1[i]:
                mismatch1 += 1
            if s[i] != alt2[i]:
                mismatch2 += 1

        # Sliding window approach
        for i in range(n, 2 * n):
            min_flips = min(min_flips, mismatch1, mismatch2)  # Update minimum flips
            
            # Remove the leftmost character (out of window)
            if s[i - n] != alt1[i - n]:
                mismatch1 -= 1
            if s[i - n] != alt2[i - n]:
                mismatch2 -= 1

            # Add the new rightmost character (entering the window)
            if s[i] != alt1[i]:
                mismatch1 += 1
            if s[i] != alt2[i]:
                mismatch2 += 1

        # Final update after last window shift
        min_flips = min(min_flips, mismatch1, mismatch2)

        return min_flips
```
