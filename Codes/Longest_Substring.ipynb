```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if len(s) == 0:
            return 0

        max_word_length = 0
        word_set = set()
        i = 0

        for j in range(len(s)):
            while s[j] in word_set:
                word_set.remove(s[i])
                i += 1
            word_set.add(s[j])
            max_word_length = max(max_word_length, j - i + 1)

        return max_word_length
```
