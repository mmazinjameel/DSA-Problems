```python
class Solution:
    def removeKdigits(self, num: str, k: int) -> str:
            stack = []

            for element in num:
                while stack and k>0 and int(stack[-1]) > int(element):
                    stack.pop()
                    k = k-1

                stack.append(element)

            ## Now we have elements in increasing order
            ## If K is still left, pop last elements
            while(k>0):
                stack.pop()
                k-=1

            res = "".join(stack).lstrip('0')
            return res if res else "0"


```
