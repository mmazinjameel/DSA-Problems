```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:

        if len(digits)==0:
            return []

        letter_map = {
            '2': 'abc',
            '3': 'def',
            '4': 'ghi',
            '5': 'jkl',
            '6': 'mno',
            '7': 'pqrs',
            '8': 'tuv',
            '9': 'wxyz'
        }

        ans, sol = [],[]
        n = len(digits)

        def backtrack(i=0):
            if i==n:
                ans.append(''.join(sol))
                return

            for letter in letter_map[digits[i]]:
                sol.append(letter)
                backtrack(i+1)
                sol.pop()


        backtrack()
        return ans
```
