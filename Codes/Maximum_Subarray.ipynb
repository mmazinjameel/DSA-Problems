```python
class Solution(object):
    def maxSubArray(self, nums):

        max_sum = float('-inf')
        cur_sum = 0

        if len(nums) == 1:
            return nums[0]

        for i in range(len(nums)):

            cur_sum += nums[i]
            
            max_sum = max(max_sum,cur_sum)
            cur_sum = max(cur_sum, 0)

        return max_sum
```
