```python
class Solution:
    def totalFruit(self, fruits: List[int]) -> int:

        basket = defaultdict(int)
        left = 0
        maxFruits = 0

        for right in range(len(fruits)):
            # Add the current fruit to the basket
            basket[fruits[right]] += 1

            # If the basket contains more than 2 types of fruits, shrink the window
            while len(basket) > 2:
                if basket[fruits[left]] == 1:
                    del basket[fruits[left]]  # Remove the fruit type completely if count is 1
                else:
                    basket[fruits[left]] -= 1  # Reduce the count of the leftmost fruit

                left += 1  # Move the left pointer forward

            maxFruits = max(maxFruits, right - left + 1)

        return maxFruits
```
