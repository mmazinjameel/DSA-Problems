# 2D Dynamic Programming

#### Start from the last column, and traverse up to down

```python
class Solution:
    def maxGold(self, mat):
        m = len(mat)
        n = len(mat[0])
        
        # Create a DP matrix
        dp = [[0 for _ in range(n)] for _ in range(m)]
        
        # Traverse from right to left
        for col in range(n - 1, -1, -1):
            for row in range(m):
                # Move right
                right = dp[row][col + 1] if col < n - 1 else 0
                # Move right-up
                right_up = dp[row - 1][col + 1] if row > 0 and col < n - 1 else 0
                # Move right-down
                right_down = dp[row + 1][col + 1] if row < m - 1 and col < n - 1 else 0
                
                # Update dp[row][col]
                dp[row][col] = mat[row][col] + max(right, right_up, right_down)
        
        # Return the max value from the first column (any row can be the starting point)
        return max(dp[i][0] for i in range(m))
```
