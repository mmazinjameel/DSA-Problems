```python
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:

        # Step 1: Reverse the list
        p = head
        q = None
        r = None

        while p:
            r = q
            q = p
            p = p.next
            q.next = r

        # Now q is the head of the reversed list
        rev_head = q
        p = rev_head
        prev = None

        # Step 2: Traverse to the nth node from beginning of reversed list
        for i in range(1, n):
            prev = p
            p = p.next

        # Step 3: Delete the nth node from the reversed list
        if prev is None:
            # Removing the first node (which was the nth from the end in original list)
            rev_head = p.next
        else:
            prev.next = p.next

        # Step 4: Reverse the list back to original order
        p = rev_head
        q = None
        r = None

        while p:
            r = q
            q = p
            p = p.next
            q.next = r

        return q  # q is the head of the final list
```
