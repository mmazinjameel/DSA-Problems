```python
class Solution:
  def countServers(self, grid: list[list[int]]) -> int:

        m = len(grid)
        n = len(grid[0])

        row_count = [0] * m
        col_count = [0] * n


        # First pass: Count no of servers in each row and column
        for i in range(m):
            for j in range(n):
                if grid[i][j]==1:
                    row_count[i]+=1
                    col_count[j]+=1

        # Second pass: Count no of servers that can communicate (basically that value in row_count or col_count should be >1)

        servers = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j]==1 and (row_count[i] > 1 or col_count[j] >1):
                    servers +=1

        return servers                
                
```
