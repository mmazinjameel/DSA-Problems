```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNodes(self, head: Optional[ListNode]) -> Optional[ListNode]:

        stack = []  # Stack to maintain decreasing values
        current = head
        while current:
            # Remove elements from the stack that are smaller than the current value
            while stack and current.val > stack[-1]:
                stack.pop()
            stack.append(current.val)
            current = current.next

        DummyNode = ListNode(stack[0])
        last = DummyNode

        for i in range(1,len(stack)):
            t = ListNode(stack[i])
            last.next = t
            last = t

        return DummyNode
```
