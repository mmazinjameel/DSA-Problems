```python
def maxAlternatingSum(nums) -> int:
    
    def helper(i=0, even=True):
        if i == len(nums):
            return 0

        # Skip current element
        skip = helper(i + 1, even)

        # Include current element
        if even:
            take = nums[i] + helper(i + 1, False)
        else:
            take = -nums[i] + helper(i + 1, True)

        return max(skip, take)

    return helper()
```

# Memoization

```python
class Solution:
    def maxAlternatingSum(self, nums: List[int]) -> int:
        memo = {}

        def helper(i=0, even=True):
            key = (i, even)
            if key in memo:
                return memo[key]

            if i == len(nums):
                return 0

            # Option 1: skip current element
            skip = helper(i + 1, even)

            # Option 2: include current element
            if even:
                take = nums[i] + helper(i + 1, False)
            else:
                take = -nums[i] + helper(i + 1, True)

            memo[key] = max(skip, take)
            return memo[key]

        return helper()
```

# Dynamic Programming

```python
class Solution:
    def maxAlternatingSum(self, nums: List[int]) -> int:
        sumEven, sumOdd = 0,0

        for i in range(len(nums)-1,-1,-1):
            tmpEven = max(sumOdd + nums[i], sumEven)
            tmpOdd = max(sumEven - nums[i], sumOdd)

            sumEven, sumOdd = tmpEven, tmpOdd

        return sumEven
```
