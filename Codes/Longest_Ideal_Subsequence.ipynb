```python
class Solution:
    def longestIdealString(self, s: str, k: int) -> int:

        dp = [0] * 26

        for ch in s:
            idx = ord(ch) - ord('a')
            max_len = 0

            start = max(0, idx-k)
            end = min(25,idx+k)

            for i in range(start,end+1):
                max_len = max(max_len, dp[i])

            dp[idx] = max(dp[idx], max_len+1)

        return max(dp)
```
