```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:

        ans, sol = [], []

        def comb(target, j=0):
            for i in range(j, len(candidates)):
                difference = target - candidates[i]
                
                if difference > 0:
                    sol.append(candidates[i])
                    comb(target=difference, j=i) 
                    sol.pop()  
                elif difference == 0:
                    sol.append(candidates[i])
                    ans.append(sol.copy())
                    sol.pop()  

        comb(target)
        return ans
```
