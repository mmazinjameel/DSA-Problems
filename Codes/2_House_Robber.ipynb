```python
class Solution:
    def rob(self, nums: List[int]) -> int:

        loot = []

        for i in range(len(nums)):
            if i==0:
                loot.append(nums[i])

            elif i==1:
              loot.append(max(nums[i],nums[i-1]))

            else:
                loot.append(max(loot[i-2]+nums[i],loot[i-1]))

        return max(loot)
```
