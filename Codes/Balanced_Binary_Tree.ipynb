```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isBalanced(self, root: Optional[TreeNode]) -> bool:

        def height_and_balance(root):
            if root is None:
                return True, 0  # (isBalanced, height)

            left_balanced, left_height = height_and_balance(root.left)
            right_balanced, right_height = height_and_balance(root.right)

            # Check if the current node is balanced
            current_balanced = left_balanced and right_balanced and abs(left_height - right_height) <= 1
            
            # Current height is the maximum of the left and right subtree heights, plus 1 for the current node
            current_height = max(left_height, right_height) + 1
            
            return current_balanced, current_height

        is_balanced, _ = height_and_balance(root)
        return is_balanced
```
