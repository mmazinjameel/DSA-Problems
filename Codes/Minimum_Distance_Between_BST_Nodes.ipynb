```python
class Solution:
    def minDiffInBST(self, root: Optional[TreeNode]) -> int:
        self.prev = None  # To store the previous node in in-order traversal
        self.min_diff = float('inf')  # Initialize min difference as infinity


        def helper(node):

            if not node:
                return

            ## Inorder Traversal
            helper(node.left)

            if self.prev is not None:
                self.min_diff = min(self.min_diff, node.val - self.prev)

            self.prev = node.val


            helper(node.right)


        helper(root)
        return self.min_diff
```
