```python
class Solution:
    def isValid(self, s: str) -> bool:
        hashmap = {
            ')':'(',
            ']':'[',
            '}':'{',
        }

        stack = []

        for element in s:
            if element in hashmap.values():  # If it's an opening bracket
                stack.append(element)
            elif element in hashmap.keys():  # If it's a closing bracket
                if not stack or stack[-1] != hashmap[element]:  # Check if stack is empty or top of stack doesn't match
                    return False
                stack.pop()

        return len(stack) == 0  # Stack should be empty at the end
```
