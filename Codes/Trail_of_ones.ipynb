# Recursion

class Solution:
    def countConsec(self, n: int) -> int:
        # code here n = 2

        def helper(string = '', consecutive_ones = False):
            if len(string) == n:
                return 0 if consecutive_ones else 1
            
            # Include the 0 at the end
            zero = helper(string = string + '0', consecutive_ones = consecutive_ones)
            
            # Include the 1 at the end
            # If last is 1 then True
            if string and string[-1] == '1':
                one = helper(string = string + '1', consecutive_ones = True)
            else:
                one = helper(string = string + '1', consecutive_ones = consecutive_ones)
        
            return zero + one
        
        return((2**n) - helper())

        

# Dynamic Programming

- a[i]: The number of valid binary strings of length i that end with 0.
    - You can append a 0 to any valid string of length i-1, whether it ends with 0 or 1.

- b[i]: The number of valid binary strings of length i that end with 1.
    - You can append a 1 only to strings of length i-1 that end with 0  to avoid having consecutive 1s.

- dp[i]: The total number of valid binary strings of length i with no consecutive 1s.

```python
class Solution:
    def countConsec(self, n: int) -> int:
        # code here n = 2

        a = [0] * (n + 1)
        b = [0] * (n + 1)
        dp = [0] * (n + 1)
        
        a[1] = 1  # "0"
        b[1] = 1  # "1"
        dp[1] = 2
        
        for i in range(2, n + 1):
            a[i] = a[i - 1] + b[i - 1]
            b[i] = a[i - 1]
            dp[i] = a[i] + b[i]

        return((2**n) - dp[n])

         
```
