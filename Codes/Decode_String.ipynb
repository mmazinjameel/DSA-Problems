```python
class Solution:
    def decodeString(self, s: str) -> str:

        stack = []
        current_string = ""
        current_number = 0

        for val in s:
            if val.isdigit(): ## If it's a digit, keep track of it
                current_number = current_number * 10 + int(val)

            elif val == '[':   # Found the opening bracket -> string starts
                stack.append((current_string,current_number))
                current_string = ""
                current_number = 0

            elif val == ']':  # Found the complete string
                prev_string , repeat_count = stack.pop()
                current_string = prev_string + current_string * repeat_count

            else:
                current_string += val
            

        return current_string
```
