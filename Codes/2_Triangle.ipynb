```python
class Solution(object):
    def minimumTotal(self, triangle):
        ans_triangle = []


        for i in range(len(triangle)):
            j = 0
            ans_triangle_level = []

            if i==0:
                ans_triangle_level.append(triangle[i][0])

            else:
                while j<len(triangle[i]):
                    if j==0:
                        ans_triangle_level.append(ans_triangle[i-1][0] + triangle[i][j])

                    elif j==len(triangle[i])-1:
    
                        ans_triangle_level.append(ans_triangle[i-1][j-1] + triangle[i][j])

                    else:
                        min_sum = min(ans_triangle[i-1][j-1] + triangle[i][j],triangle[i][j]+ ans_triangle[i-1][j])
                        ans_triangle_level.append(min_sum)

                    j+=1

            ans_triangle.append(ans_triangle_level)
   

        last_ptr = len(ans_triangle)-1
        minimum_path = min(ans_triangle[last_ptr])
        return minimum_path
                
```
