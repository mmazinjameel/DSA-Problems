```python
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        
        ## Start with max_jump = 0
        ## Increase the counter with max_jump = max(max_jump,nums[i]+i)
        ## While traversing:
        ##      if nums[i] + i >= len(nums) and max_jump >= i:
        ##          return True
        ##      0 case:
        ##          if max_jump == i:
        ##          return False
        ##      keep on updating max_jump
        
        max_jump = 0
        for i in range(len(nums)):
            if nums[i] + i >= len(nums) - 1 and max_jump >= i: 
                return True
            if nums[i] == 0 and max_jump == i:  
                return False
            max_jump = max(max_jump, nums[i] + i)
```
