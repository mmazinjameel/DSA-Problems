```python
class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:
        
        row = len(grid)
        col = len(grid[0])
        per = 0

        for i in range(row):
            for j in range(col):
                if grid[i][j] == 1:
                    per += 4
                    # up
                    if i - 1 >= 0 and grid[i - 1][j] == 1:
                        per -= 1
                    # left
                    if j - 1 >= 0 and grid[i][j - 1] == 1:
                        per -= 1
                    # right
                    if j + 1 < col and grid[i][j + 1] == 1:
                        per -= 1
                    # down
                    if i + 1 < row and grid[i + 1][j] == 1:
                        per -= 1

        return per
```
