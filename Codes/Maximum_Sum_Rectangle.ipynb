```python
class Solution:
    def maxRectSum(self, mat):
        def Kadanes(nums):
            max_sum = nums[0]
            cur_sum = nums[0]
            for n in nums[1:]:
                cur_sum = max(n, cur_sum + n)
                max_sum = max(max_sum, cur_sum)
            return max_sum

        rows = len(mat)
        cols = len(mat[0])
        max_rect_sum = float('-inf')

        for top in range(rows):
            # Initialize a temporary array to accumulate column sums
            arr = [0] * cols

            for bottom in range(top, rows):
                # Add the current row to the temp array
                for col in range(cols):
                    arr[col] += mat[bottom][col]

                # Apply Kadane's on the collapsed array
                current_max = Kadanes(arr)
                max_rect_sum = max(max_rect_sum, current_max)

        return max_rect_sum
```
