```python
class Solution:
    def balanceSums(self, mat):
        row_sum = [sum(row) for row in mat]
        col_sum = [sum(col) for col in zip(*mat)]
        
        max_sum = max(max(row_sum), max(col_sum))
        
        operations = 0
        
        for i in range(len(row_sum)):
            operations += max_sum - row_sum[i] 
            
            
        return operations
```
