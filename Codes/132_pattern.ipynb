```python
class Solution:
    def find132pattern(self, nums: List[int]) -> bool:
        stack = []  # pair [num, minLeft]
        curMin = nums[0]

        for n in nums[1:]:
            while stack and n>= stack[-1][0]:   # Removing any bigger digit (because we want to maintain    monotonically decreasing stack)
                stack.pop()

            # if found a min number and stack, then we have to check.. if it is greater than the min also
            if stack and n < stack[-1][0] and n > stack[-1][1]:
                return True

            stack.append([n,curMin])
            curMin = min(curMin, n)

        return False
```
