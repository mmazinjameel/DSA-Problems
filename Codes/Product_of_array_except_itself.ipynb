```python

class Solution:
    def productExceptSelf(self, nums):
        output = [1] * len(nums)

        # Calculate left products and store in output
        left_product = 1
        for i in range(len(nums)):
            output[i] = left_product
            left_product *= nums[i]

        # Calculate right products and multiply with output
        right_product = 1
        for i in range(len(nums) - 1, -1, -1):
            output[i] *= right_product
            right_product *= nums[i]

        return output

```
