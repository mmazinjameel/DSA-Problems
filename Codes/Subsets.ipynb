```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        def subsets(level, current_set, res):
            if level == len(nums):  # Base case: if we've reached the end of nums
                res.append(current_set[:])  # Append a copy of the current_set to results
                return
            
            # Include the current element in the subset
            current_set.append(nums[level])
            subsets(level + 1, current_set, res)
            
            # Exclude the current element from the subset
            current_set.pop()
            subsets(level + 1, current_set, res)

        res = []
        subsets(0, [], res)
        return res
        
```
