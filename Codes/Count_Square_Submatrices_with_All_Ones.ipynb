```python
class Solution:
    def countSquares(self, matrix: List[List[int]]) -> int:
        if not matrix or not matrix[0]:
            return 0

        rows, cols = len(matrix), len(matrix[0])
        dp = [[0] * cols for _ in range(rows)]
        total_squares = 0

        for i in range(rows):
            for j in range(cols):
                if matrix[i][j] == 1:  # matrix contains integers 0 or 1
                    if i == 0 or j == 0:
                        dp[i][j] = 1
                    else:
                        dp[i][j] = 1 + min(
                            dp[i-1][j],      # top
                            dp[i][j-1],      # left
                            dp[i-1][j-1]     # top-left
                        )
                    total_squares += dp[i][j]  # add current cell's count to total

        return total_squares
```
